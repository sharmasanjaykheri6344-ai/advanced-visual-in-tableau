Q.1  Create a Heatmap to analyze Treatment Cost by Department and Region?

...>> ✅ Step-by-Step: Create Heatmap in Tableau
🎯 Business Goal

Management wants to identify:

Which Departments are expensive

Which Regions incur higher treatment costs

Where cost optimization is required 💰

⭐ Step 1 — Open a New Worksheet

Load your dataset → Go to Worksheet

⭐ Step 2 — Place Dimensions

Drag Department → Rows

Drag Region → Columns

👉 This creates a grid structure for the heatmap.

⭐ Step 3 — Add Measure

Drag Treatment Cost → Color (Marks Card)

Now Tableau will automatically create a heatmap using color shading.

⭐ Step 4 — Improve Visualization

Make the heatmap clearer:

Change Marks Type → Square

Drag Treatment Cost → Label (optional for exact values)

Click Color → Edit Colors

Use Red (High Cost)

Use Light Yellow / Green (Low Cost)

📊 Business Insights You Can Derive
✅ Identify Costly Departments

Example:

Surgery → Dark Red → High treatment cost

General Medicine → Light color → Low cost

👉 Helps hospital management review:

Equipment cost

Specialist charges

Process efficiency

✅ Regional Cost Comparison

If North Region shows darker shades, it may indicate:

Higher operational cost

Expensive vendors

Higher patient complexity

👉 Supports budgeting and pricing decisions.

✅ Resource Allocation

Management can:

Allocate more budget to high-cost departments

Improve efficiency in specific regions

Standardize treatment protocols 📈

⭐ Pro Tip (Very Important for Assignments)

Instead of SUM(Treatment Cost), try:

AVG(Treatment Cost) → Understand typical cost

COUNT(Patient ID) → Volume vs cost comparison

This makes the heatmap more analytical rather than just descriptive 👍

Q.3 Create a Bump Chart to show ranking of Departments based on Treatment Cost over months?

...>> To visualize how Department rankings change over months based on Treatment Cost, you can create a Bump Chart in Tableau.
A bump chart is excellent for showing rank movement (who moves up or down) over time 📈

✅ Step-by-Step: Create Bump Chart in Tableau
🎯 Business Goal

Management wants to know:

Which departments are becoming more expensive over time

Which departments are improving cost efficiency

Trend in cost ranking month-wise 💰

⭐ Step 1 — Create Month Field (if needed)

If Order/Treatment date exists:

Drag Treatment Date → Columns

Change to Month (Discrete)

⭐ Step 2 — Add Department

Drag Department → Color (Marks Card)
This creates separate lines for each department.

⭐ Step 3 — Create Ranking Calculation

Create Calculated Field → Department Cost Rank

RANK(SUM([Treatment Cost]))
⭐ Step 4 — Build the Bump Chart

Drag Department Cost Rank → Rows

Right click axis → Reverse Axis
👉 Rank 1 should appear at the top.

⭐ Step 5 — Convert to Line Chart

Change Marks Type → Line

Drag Department → Label (optional for clarity)

⭐ Step 6 — Set Table Calculation Direction

Right click Rank field → Edit Table Calculation

Compute Using → Department

Restart every → Month

👉 This ensures ranking is recalculated for each month.

📊 Insights Management Can Derive
✅ Identify Rising Cost Departments

If Surgery rank moves from 4 → 1, it means:

Treatment costs are increasing rapidly

Requires cost audit or vendor review

✅ Detect Improving Efficiency

If Radiology rank drops from 1 → 3:

Indicates better equipment utilization

Improved operational efficiency

✅ Budget Planning & Resource Allocation

Leadership can:

Predict future cost pressure

Adjust department budgets

Plan staffing and infrastructure

⭐ Pro Tip (Very Powerful)

Add:

Tooltip → SUM(Treatment Cost)

Filter → Region / Hospital Branch

Parameter → Top N Departments

This converts a simple bump chart into an executive-level analytical visual 🚀

Q.4  Create a Butterfly Chart comparing Male vs Female patient count by Department.

...>> ✅ Business Goal

Management wants to understand:

Gender distribution across departments

Which departments have higher male or female patient load

Support staffing, facilities, and service planning

✅ Step-by-Step: Create Butterfly Chart in Tableau
⭐ Step 1 — Create Patient Count Measure

If Patient_ID exists:

Create Calculated Field → Patient Count

COUNT([Patient_ID])
⭐ Step 2 — Create Male Patient Count

Create Calculated Field → Male Count

IF [Gender] = "Male" THEN 1 END

Then use:

SUM([Male Count])
⭐ Step 3 — Create Female Patient Count

Create Calculated Field → Female Count

IF [Gender] = "Female" THEN 1 END

Then use:

SUM([Female Count])
⭐ Step 4 — Make One Side Negative (Very Important)

Create field → Female Count (Negative)

- SUM([Female Count])

👉 This pushes female bars to the left side, creating butterfly effect 🦋

⭐ Step 5 — Build the Chart

Drag Department → Rows

Drag Female Count (Negative) → Columns

Drag Male Count → Columns (next to it)

⭐ Step 6 — Convert to Dual Axis

Right click second axis → Dual Axis

Synchronize Axis

Change Marks → Bar

⭐ Step 7 — Improve Formatting

Color Male → Blue

Color Female → Pink

Show labels → Use ABS() for female values so labels don’t show negative

📊 Insights Management Can Derive
✅ Gender Load by Department

Example:

Orthopedics → Higher Male patients

Gynecology → Higher Female patients

👉 Helps allocate:

Specialists

Nursing staff

Equipment

✅ Infrastructure Planning

If Pediatrics has balanced gender load:

Requires neutral resource allocation

If Cardiology shows heavy male dominance:

Awareness programs can be planned for females ❤️

✅ Service Strategy

Customize treatment packages

Improve targeted marketing

Enhance patient satisfaction

⭐ Pro Tip (Advanced Dashboard Feature)

Add:

Filter → Region / Month

Tooltip → Avg Treatment Cost

Parameter → Top N Departments

This turns butterfly chart into a strategic healthcare planning visual 🚀

Q.5  Create a Dual Axis Chart to compare Average Age and Treatment Cost by Department.

...>> ✅ Business Goal

Management wants to understand:

Which departments treat older vs younger patients

Support pricing strategy and specialist allocation 👩‍⚕️

✅ Step-by-Step: Create Dual Axis Chart
⭐ Step 1 — Place Dimension

Drag Department → Columns

⭐ Step 2 — Add First Measure (Average Treatment Cost)

Drag Treatment Cost → Rows

Right click → Measure → Average

⭐ Step 3 — Add Second Measure (Average Age)

Drag Age → Rows (next to first measure)

⭐ Step 4 — Convert to Dual Axis

👉 Now both measures appear in the same chart.

⭐ Step 5 — Change Chart Types (Best Practice)

Set Treatment Cost → Bar

This improves readability 📈

⭐ Step 6 — Improve Formatting

Q.6  Create a Sunburst Chart showing Department → Diagnosis hierarchy by Treatment Cost?

...>> ✅ Business Goal

Management wants to know:

Which Departments generate highest treatment cost

Within each department, which Diagnosis contributes most

Identify high-cost disease areas for budgeting and optimization 💰

✅ Step-by-Step: Create Sunburst Chart in Tableau

(Sunburst is created using Pie + Dual Axis technique)

⭐ Step 1 — Create Hierarchy

In Data Pane → Right click Department → Create Hierarchy

Drag Diagnosis into that hierarchy

👉 Now you have:

Department → Diagnosis hierarchy

⭐ Step 2 — Create Total Treatment Cost

Use existing measure:

SUM(Treatment Cost)

⭐ Step 3 — Build Inner Ring (Department Level)

Drag MIN(1) → Columns

Change Marks → Pie

Drag Department → Color

Drag SUM(Treatment Cost) → Angle

Increase Size

⭐ Step 4 — Build Outer Ring (Diagnosis Level)

Drag another MIN(1) → Columns

Make Dual Axis

Synchronize Axis

On second Marks card:

Drag Diagnosis → Color

Drag SUM(Treatment Cost) → Angle

Increase Size (larger than inner)

👉 Now outer ring represents Diagnosis contribution

⭐ Step 5 — Format Sunburst

Remove headers

Add borders between slices

Adjust colors for hierarchy clarity

Add tooltip:

Department

Diagnosis

Treatment Cost

% Contribution

📊 Insights Management Can Derive
✅ Identify Costliest Departments

Example:

Surgery → Largest inner slice

👉 Needs budget monitoring and efficiency programs.

✅ Detect High-Cost Diagnoses

Inside Surgery:

Cardiac Surgery → Major outer slice

👉 Helps:

Specialist hiring

Equipment investment

Insurance pricing decisions ❤️

✅ Strategic Healthcare Planning

Leadership can:

Focus preventive care on high-cost diseases

Optimize treatment protocols

Improve procurement strategy

⭐ Pro Tip (Very Important)

Add:

Filter → Region / Month

Parameter → Top N Diagnoses

Combine with Bump Chart or Heatmap

👉 This creates a complete cost-analysis storytelling dashboard 🚀
